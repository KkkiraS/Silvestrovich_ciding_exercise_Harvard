```python
def get_personal_page(paragraphs):
    x = paragraphs[1].split('/')
    personal = x[-2].split('"')
    personal_page = ('/').join([base_url,personal[0]])
    return personal_page
```

#### Publications

```python
publications = soup2.find_all('div', class_='field field--name-field-external-publications field--type-link field--label-above')
```

```python
def get_publication(publications):
    if len(publications) == 0:
        publication_site = ''
    else:
        draft = publications[0].find_all('a')
        publication_site = str(draft[0]).split('"')[1]
    
    return publication_site
    
```

Now let's have our final function

```python
people_list = []
for i in range(43):
    
    p = people[i].find_all('a')
    paragraphs = []
    for x in p:
        paragraphs.append(str(x))
    personal_list = []
    personal_list.append(people[i].get_text().strip().split('\n')[0])
    personal_list.append(get_personal_page(paragraphs))
    personal_list.append(get_personal_website(paragraphs))
    
    base_url_2 = get_personal_page(paragraphs)
    base_page2 = requests.get(base_url_2)
    soup2 = BeautifulSoup(base_page2.content, 'html.parser')
    
    campus_affiliation = soup2.find_all('div', class_='field field--name-field-campus-affiliations field--type-string field--label-above')
    offcampus_affiliation = soup2.find_all('div', class_='field field--name-field-off-campus-affiliations field--type-string field--label-above')
    publications = soup2.find_all('div', class_='field field--name-field-external-publications field--type-link field--label-above')
    
    campus = get_affiliation(campus_affiliation)
    offcampus = get_affiliation(offcampus_affiliation)
    
    if campus == '': 
        campus_length = 0
    else:
        campus_length = len(campus.split(' ; '))

    if offcampus == '': 
        offcampus_length = 0
    else:
        offcampus_length = len(offcampus.split(' ; '))
    
    personal_list.append(campus_length)
    personal_list.append(campus)
    
    personal_list.append(offcampus_length)
    personal_list.append(offcampus)
    
    personal_list.append(get_publication(publications))
    
    people_list.append(personal_list)

people_list
```

Then, I collect data on a) name of publications b)years of these publications, count number of publications in each year for each person who has google scholar page and merge it into one dataframe

```python
frames = []
for i in range(len(df)):
    name = df.loc[i,'Name']
    google_scholar = df.loc[i,'google_scholar']
    if google_scholar != '':
        publication_list = []
        for start in range(0, 1600, 100):
            r = requests.get(
                ('').join([google_scholar,'&view_op=list_works&sortby=year&cstart={}&pagesize=100']).format(start))
            soup = BeautifulSoup(r.text, features="html.parser")
            for publication,year in zip(soup.findAll('a', attrs={'class': 'gsc_a_at'}),soup.findAll('span', attrs={'class': 'gs_oph'})):
                publication_list.append([name,publication.text,int(year.text.strip(', '))])
        df_pub = pd.DataFrame(publication_list)
        df_pub.rename(columns={0: "Name", 1: "Publication", 2: "year"}, errors="raise", inplace = True)
        right = df_pub.groupby(by = ['Name','year']).count().query('year > 2000')
        right.reset_index(level = [0,1], inplace = True)
        frames.append(right)
    else:
        right = pd.DataFrame(columns = [0,1,2])
        right.rename(columns={0: "Name", 1: "Publication", 2: "year"}, errors="raise", inplace = True)
        frames.append(right)
        
right_df = pd.concat(frames)
```

```python
Harvard_predoc_with_publications = df.merge(right_df, on='Name', how='left')
```
